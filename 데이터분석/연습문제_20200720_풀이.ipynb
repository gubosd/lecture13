1. 아이리스 데이터에서 setosa 품종의 네가지 속성의 평균값을
   리스트로 구하시오. a.mean() 함수 사용
2. 나머지 두 품종에 대해서도 같은 것을 구하시오.
3. 1번과 2번 문제의 결과를 꺽은선 그래프 3개로 그리시오. plt.plot()

```python
iris_labels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']

iris = np.loadtxt('iris.csv', skiprows=1, delimiter=',',
                  converters={4: lambda s: iris_labels.index(s.decode())})
```

```python
setosa = iris[:50]
versicolor = iris[50:100]
virginica = iris[100:]
```

```python
setosa.shape
```

```python
m1 = [setosa[:,0].mean(), setosa[:,1].mean(), setosa[:,2].mean(),
      setosa[:,3].mean()]
m1
```

```python
m1 = [setosa[:,i].mean() for i in range(4)]
m1
```

```python
m2 = [versicolor[:,0].mean(), versicolor[:,1].mean(), versicolor[:,2].mean(),
      versicolor[:,3].mean()]
m2
```

```python
m3 = [virginica[:,0].mean(), virginica[:,1].mean(), virginica[:,2].mean(),
      virginica[:,3].mean()]
m3
```

```python
plt.plot(m1)
plt.plot(m2)
plt.plot(m3)
plt.legend(['setosa', 'versicolor', 'virginica'])
```
